One of the most famous datasets in *Machine Learning* is the _Iris Dataset_. The Iris Flower Dataset is also known as the Fisher's Iris Dataset. Sir Ranold Fisher was born in 1890 in London, England. He is described as the genius who created foundations for modern statistical science. The dataset was used in his 1936 paper *The Use of Multiple Measurements in Taxonomic Problems* - it includes three Iris species with 50 samples each, as well as some properties about each flower. At the start I would like to show what the Iris flower looks like:

```python
iris_chrysographes = imread("iris_purple_black.jpg")

plt.figure(figsize = (3, 3))

plt.xticks([])
plt.yticks([])

plt.title("Iris Chrysographes / puple-black, gold-marked", size = 8)

plt.imshow(iris_chrysographes)
```

This is the Iris Chrysographes in puple-black and gold-marked. One of the most beautiful colors ever existed and quite unique among the Iris flowers. However, there are three other Iris species used in The Fisher's Iris Dataset - *Setosa*, *Versicolor* and *Virginica*. There are 4 measurements for each flower:
- Sepal length
- Sepal width
- Petal length
- Petal width

And here I have a nice drawing for people like me who are not so much into flower details:

Now I load the Iris Data - so famous, it is included in the *SciKit-learn* (sklearn) Python library - and then I make a DataFrame for more clarity.

```python
iris = load_iris()
```

```python
iris
```

```python
iris_df = pd.DataFrame(iris.data, columns = iris.feature_names)
iris_df["target"] = iris.target

iris_df.sample(7)  # I choose samples to have displayed all three (0, 1, 2) target numbers from the DataFrame
# if you do not see all three targets, please, run again
```

*Target* is the numerical representation of the Iris flowers classes:
- Setosa is 0
- Versicolor is 1
- Virginica is 2

And yes, computers start to count from 0, humans start to count from 1. The language of machines and computers are numbers, so better we convert text data into numerical data. Besides, machine learning algorithms rely on mathematical models to make predictions.

```python
iris["target_names"][0] == "setosa"
iris["target_names"][1] == "versicolor"
iris["target_names"][2] == "virginica"  # above human explanation of the same thing, here computer says True
```

I see what the name of the first column is:

```python
iris.feature_names[0]
```

```python
sepal_length = iris.feature_names[0]   # name of the column, not the data in it
```

And the name of the second column:

```python
iris.feature_names[1]
```

```python
sepal_width = iris.feature_names[1]   # name of the column, not the data in it
```

```python
plt.figure(figsize = (10, 5))

plt.scatter(sepal_length, sepal_width, facecolor = "black", s = 25)

plt.title("Everything starts from a dot (or seed)", size = 9, loc = "right")

plt.show()
```

A scatter plot with just one dot. We all have a starting point - flowers, humans and machines. Data is the relationship between numerical variables and attributes but everything starts from a single dot. The scatter plots are diagrams where each value in the dataset is represented by a dot. So, more data - more dots. But we have a really strong data relationship if the dots are following themselves.

> Now I make a scatter plot with all data for the Sepal measurements - length and width:

```python
plt.figure(figsize = (9, 6))

x = iris_df["sepal length (cm)"]
y = iris_df["sepal width (cm)"]

plt.scatter(x, y, facecolor = "black", s = 15)

plt.xlabel("Sepal length (cm)", loc = "left")
plt.ylabel("Sepal width (cm)", loc = "bottom")

plt.title("Iris Flower Dataset, sepals", size = 9, loc = "right")

plt.show()
```

>Here I do the same for the Petals - I display the columns for my clarity and then I show the data in a scatter plot:

```python
iris_df[["petal length (cm)", "petal width (cm)"]]
```

```python
plt.figure(figsize = (9, 6))

x = iris_df["petal length (cm)"]
y = iris_df["petal width (cm)"]

plt.scatter(x, y, facecolor = "black", s = 15)

plt.xlabel("Petal length (cm)", loc = "left")
plt.ylabel("Petal width (cm)", loc = "bottom")

plt.title("Iris Flower Dataset, petals", size = 9, loc = "right")

plt.show()
```

>> You see how many dots we have with more data? Scatter plots equal life and knowledge.

The Sepal and Petal scatter plots are showing the data for all three Iris species but let us find out whose data is who's and how they relate to each other:

```python
plt.figure(figsize = (18, 5))

rows = 1
columns = 3

# Iris Setosa

plt.subplot(rows, columns, 1)

plt.xlim(4, 8.2)
plt.ylim(1.8, 4.7)

x_setosa = iris_df[iris_df["target"] == 0]["sepal length (cm)"]
y_setosa = iris_df[iris_df["target"] == 0]["sepal width (cm)"]

plt.scatter(x_setosa, y_setosa, facecolor = "red", s = 12)

plt.xlabel("Sepal length (cm)", loc = "left")
plt.ylabel("Sepal width (cm)", loc = "bottom")

plt.title("Iris Setosa, sepals", loc = "right")

# Iris Versicolor

plt.subplot(rows, columns, 2)

plt.xlim(4, 8.2)
plt.ylim(1.8, 4.7)

x_versicolor = iris_df[iris_df["target"] == 1]["sepal length (cm)"]
y_versicolor = iris_df[iris_df["target"] == 1]["sepal width (cm)"]

plt.scatter(x_versicolor, y_versicolor, facecolor = "green", s = 12)

plt.xlabel("Sepal length (cm)", loc = "left")
plt.ylabel("Sepal width (cm)", loc = "bottom")

plt.title("Iris Versicolor, sepals", loc = "right")

# Iris Virginica

plt.subplot(rows, columns, 3)

plt.xlim(4, 8.2)
plt.ylim(1.8, 4.7)

x_virginica = iris_df[iris_df["target"] == 2]["sepal length (cm)"]
y_virginica = iris_df[iris_df["target"] == 2]["sepal width (cm)"]

plt.scatter(x_virginica, y_virginica, facecolor = "blue", s = 12)

plt.xlabel("Sepal length (cm)", loc = "left")
plt.ylabel("Sepal width (cm)", loc = "bottom")

plt.title("Iris Virginica, sepals", loc = "right")

# Show

plt.show()
```

>*Iris Setosa* sepals are wide and short while *Iris Versicolor's* are thinner and longer. *Iris Virginica* is in between. And the petals?

```python
plt.figure(figsize = (18, 5))

rows = 1
columns = 3

# Iris Setosa

plt.subplot(rows, columns, 1)

plt.xlim(0.7, 7.2)
plt.ylim(-0.1, 2.7)

x_setosa = iris_df[iris_df["target"] == 0]["petal length (cm)"]
y_setosa = iris_df[iris_df["target"] == 0]["petal width (cm)"]

plt.scatter(x_setosa, y_setosa, facecolor = "red", s = 12)

plt.xlabel("Petal length (cm)", loc = "left")
plt.ylabel("Petal width (cm)", loc = "bottom")

plt.title("Iris Setosa, petals", loc = "right")

# Iris Versicolor

plt.subplot(rows, columns, 2)

plt.xlim(0.7, 7.2)
plt.ylim(-0.1, 2.7)

x_versicolor = iris_df[iris_df["target"] == 1]["petal length (cm)"]
y_versicolor = iris_df[iris_df["target"] == 1]["petal width (cm)"]

plt.scatter(x_versicolor, y_versicolor, facecolor = "green", s = 12)

plt.xlabel("Petal length (cm)", loc = "left")
plt.ylabel("Petal width (cm)", loc = "bottom")

plt.title("Iris Versicolor, petals", loc = "right")

# Iris Virginica

plt.subplot(rows, columns, 3)

plt.xlim(0.7, 7.2)
plt.ylim(-0.1, 2.7)

x_virginica = iris_df[iris_df["target"] == 2]["petal length (cm)"]
y_virginica = iris_df[iris_df["target"] == 2]["petal width (cm)"]

plt.scatter(x_virginica, y_virginica, facecolor = "blue", s = 12)

plt.xlabel("Petal length (cm)", loc = "left")
plt.ylabel("Petal width (cm)", loc = "bottom")

plt.title("Iris Virginica, petals", loc = "right")

# Show

plt.show()
```

>The petals are smaller, as you can see from the scatter dots, but they are also more upright. Please, check the pictures in the beginning again.
